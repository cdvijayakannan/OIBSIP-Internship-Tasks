# Time Series Analysis

```python
import matplotlib.pyplot as plt

#loading the datasets

data = pd.read_csv('retail_sales_dataset.csv')
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# monthly frequency
monthly_sales = data.resample('ME', on='Date').sum()

plt.figure(figsize=(9,5))
plt.plot(monthly_sales.index, monthly_sales['Total Amount'],marker='o',linestyle='-')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid(True)

plt.show()
```

# Visualization

```python
plt.figure(figsize = (8,10))
sns.boxplot(y = data['Total Amount'], x = data['Product Category'])
```

```python
#Age Demographic Factor over Visualization

age_group = pd.cut(data['Age'],bins=[0, 18, 25, 35, 45, 55, 65, 100], labels = ['Under 18', '18-25', '26-35', '36-45','46-55','56-65', 'Over 65']) 
age_demograph = data.groupby(age_group, observed = False).size()

plt.figure(figsize = (8, 7))
age_demograph.plot(kind = 'pie', autopct = '%1.1f', labels = None)
plt.title('Age Demographic')
plt.ylabel('')
plt.legend(age_demograph.index, loc = 'best')
plt.show()
```

```python
#ANALYSING THE TOP PRODUCT
top_products = data.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False)
plt.figure(figsize =(10,6))
top_products.plot(kind='barh',color='green')
plt.xlabel('Total Quantity')
plt.ylabel('Product Category')
plt.title('Top Products')
plt.show() 
```

```python
#ANALYSIS OF THE MONTHLY PURCHASE TRENDS OVER VISUALIZATION

import matplotlib.pyplot as plt
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data['Month'] = data['Date'].dt.month
monthly_purchase = data.groupby('Month').size()

plt.figure(figsize=(10, 6))
plt.plot(monthly_purchase.index, monthly_purchase.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.title('Monthly Purchases Trend')
plt.xticks(monthly_purchase.index)  # All months are in x-axis
plt.grid(True)
plt.show()
```

```python
#Sales Trends Heatmap Visualization

heatmap = data.pivot_table(index='Month', columns='Product Category', values='Total Amount')
plt.figure(figsize=(12,8))
sns.heatmap(heatmap, annot=True, fmt='.0f')
plt.title('Sales Heatmap')
plt.xlabel('Product Category')
plt.ylabel('Month')
plt.show()
```

```python
#Correlation between normalized price unit vs normalized price amount by Scatter plotting visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Loading the datasets
data = pd.read_csv('retail_sales_dataset.csv')

# Normalize the columns 'Price per Unit' and 'Total Amount' using z-score normalization

data['Price per Unit'] = (data['Price per Unit'] - data['Price per Unit'].mean()) / data['Price per Unit'].std()
data['Total Amount'] = (data['Total Amount'] - data['Total Amount'].mean()) / data['Total Amount'].std()

# Calculate the correlation between normalized 'Price per Unit' and 'Total Amount'

correlation = data['Price per Unit'].corr(data['Total Amount'])

# Plotting a scatter plot to visualize the relationship between normalized 'Price per Unit' and 'Total Amount'
plt.scatter(data['Price per Unit'], data['Total Amount'])
plt.xlabel('Normalized Price per Unit')
plt.ylabel('Normalized Total Amount')
plt.title('Scatter plot of Normalized Price per Unit vs Normalized Total Amount')
plt.show()

# Plotting a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
correlation_matrix = data[['Price per Unit', 'Total Amount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Normalized 'Price per Unit' and 'Total Amount'")
plt.show()

# Detecting outliers using z-score method on normalized data
z_scores = stats.zscore(data[['Price per Unit', 'Total Amount']])
abs_z_scores = abs(z_scores)
outlier_rows = (abs_z_scores > 3).any(axis=1)
outliers = data[outlier_rows]

print("Correlation between normalized 'Price per Unit' and normalized 'Total Amount':", correlation)
print("Outliers:")
print(outliers)
```
